```python
def find_filepath(foldername):
    label_names = [name for name in os.listdir(foldername)]
    sub_folders = [os.path.join(foldername, name) for name in label_names]

    path_list = []
    label_list = []
    for folder, label in zip(sub_folders, label_names):
        file_list = os.listdir(folder)
        for file in file_list:
            path = os.path.join(folder, file)
            path_list.append(path)
            label_list.append(label)
    return label_names, path_list, label_list
```
