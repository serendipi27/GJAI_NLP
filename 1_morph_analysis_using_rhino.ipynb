# 형태소 분석 기초 지식
## 정형 .vs. 비정형 .vs. 반정형 데이터
 - 정형 데이터
     - 즉시 분석 가능한 구조로 정형화 되어있는 데이터
     - **첫째, 일관된 표 형태로 정리되어 있을 것.**
     - **둘째, 단순한 의미의 숫자로 기록되어 있을 것.**
 - 비정형 데이터
     - 정보를 추출하기 어려운 형태로 되어있는 데이터
     - 일관된 모습으로 저장되어 있지 않거나 의미 단위 요소를 파악하기 어려운 데이터
     - **무엇이 정보의 핵심이지 결정되어 있지 않음.**
     - 예시) SNS 데이터, 고객 게시글, QR 코드, 오디오, 비디오
 - 반정형 데이터
     - 데이터의 구조나 의미를 컴퓨터가 이해할 수 있게 표시한 데이터
     - 예시) HTML: tag를 이용해 컴퓨터가 이해할 수 있도록 구조화된 데이터
     
## 형태소 분석이 필요한 한국어
 - <u>형태소</u>
     - **의미를 가진 최소 단위**
     - 문법적으로는 "더 나누었을 때 의미를 갖지 못하는 단위"까지 나갈 수 있지만, 우리는 여기까지 나가지 않음.
     - 우리는 실용적인 접근이 필요함. 심지어 "의미"라는 범위 또한 분석가가 임의로 설정할 수 있음.
 - <u>형태소 분석이란?</u>
     - 만약 어절 단위 그대로 사전에 등재하고, 이를 조회한다면, 300,000 * 1,000 * 5,6000 = 1,680,000,000,000 번 조회를 해야함. 실질적으로 자연어처리가 불가능하다고 봄.
     - 어절에서 어근을 떼어내고, 의미가 적은 접사(조사+어미)를 제외시킴.
     - 사전적 접근으로 봤을 때 컴퓨터가 처리 가능하도록 줄여줌.
     - 어근으로 따지만 약 300,000만개면 충분하며, 실제 자주 사용하는 단어는 30,000개.
     - 실 분석에서는 약 10,000개 단위로 사전에 등재하고 그 외에 들어오는 단어는 그냥 무시하는 것을 택함.
     
## 형태소 분석기? or 단어 분석기?
 - "형태소 분석기"라 쓰고 "단어 분석기"라 읽는다.
 - 영어는 두 가지 말이 동일하지만, 우리나라 말은 아니다.
 - 우리말 서술어는 단어 단위가 아닌, 형태소 단위로 분석해야 한다.
 - 그러나 복합명사에서는 그 의미가 중요하므로, 형태소 단위로 단어를 쪼개는 게 아니라, 단어 단위로 붙여서 봐야 한다.
 - **결국, 목적에 맞게 의미가 통하도록 단어나 형태소 형태로 나오면 된다.**
 
## 단어란 무엇일까?
 - 단어의 정의는 여러가지 내릴 수 있지만, 중요한 사전적 의미 중 하나는 **사전에 등재된 말**.
 - 국립국어원에서 결정함.
 - 단어의 정의는 주관적이다.

 ## XR의 중요성
  - XR: eXtended Root
  - 어근이긴 한데 품사를 받지 못한 애들을 모아 놓은 태그
  - ex) 어떨떨. 이 자체로는 쓰이지 않고 따라서 품사를 받지 못함. 그러나 얘는 "하다"가 더해지면 어떨떨하다라는 형용사가 됨. 그래서 그냥 버릴 수 없음.
  - 형태소 분석 할때 XR을 제외시키면 많은 형태소를 놓칠 수 있음.
  - 형태소 분석이 끝나고, XR을 명사류로 할지, 동사류를 할지 결정해야 함. 보통 동사류에 넣는게 타당해 보인다.

# 시동 걸기 - startRhino()
 - 형태소 분석을 위한 rhinoMorph 모듈 시작을 알리는 함수
 - 사전의 내용을 메모리에 올리는 기능
 - 밑의 예시에서는 rn에 사전을 담는 행위
 - 단 한번만 실행함.

```python
import rhinoMorph

# 시동 걸기
rn = rhinoMorph.startRhino()
text = "한글로 된 한글텍스트를 분석하는 것은 즐겁다."
```

# 형태소 분석 함수 1 - onlyMorph_list()
 - 그렇게 많이 사용되진 않는다. 왜냐하면 형태소 분석에서는 필요한 정보량을 포함한 
    명사류, 동사류 정도만 필요함.
 - 모든 형태소만 가지고 오되, list 타입으로 반환함.
 

```python
# 형태소 분석 함수 1 : 모든 형태소 보이기
text_analyzed = rhinoMorph.onlyMorph_list(rn, text) 
print('\n1. 형태소 분석 결과:', text_analyzed)
```

## 옵션 1 - pos
 - 실질형태소만, 동사의 어말어미는 제외
 - Part of Speech ==> 품사라고 번역하지만 실은 단어!!
 - 영어에서는 어떤 어근을 쓸 것인가 의미(영어는 모든 품사가 모든 어근, 단어를 포함함. 띄어쓰기로 모든 것이 구분됨. 8품사체계)
 - 우리나라에서는 실질적 의미 분석을 위해 어떤 어근만을 쓸 것인가 의미(모든 품사가 모든 어근, 단어를 포함하지 못함. 9품사체계. 어미는 9품사에 포함되지 못함. **단어와 품사의 불일치**)
 - 9품사: 명대수동형관부조감(명사, 대명사, 수사, 동사, 형용사, 관용사, 부사, 조사, 감탄사)
 - 품사태그: 9품사가 모든 단어를 아우러지 못해 나타나는 문제를 감추고, 데이터 분석을 시도하고자 나온 개념.
 - 우리나라 단어의 조건
     - 첫째, 자립할 수 있어야 한다.
     - 둘째, 단, 조사는 예외이다.

```python
# 가장 많이 사용됨
text_analyzed = rhinoMorph.onlyMorph_list(rn, text, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'])
print('\n2. 형태소 분석 결과:', text_analyzed)
```

## 옵션 2 - eomi
- 실질형태소만, 동사의 어말어미 포함
- 형용사 등의 뒤에 어미 '다'를 붙여 뜻이 통하게끔 만듦.
- 만약 "되", "즐겁" 이렇게만 형태소 분석이 되었다면 그 의미를 파악하기 어려움.
- 의뢰자가 형태소 분석이 잘못됐다고 생각할 수도 있기 때문에 조치가 필요함.
- (많이쓰임)**빈도분석에서 많이 사용**

```python
text_analyzed = rhinoMorph.onlyMorph_list(rn, text, pos=['NNG', 'NNP', 'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
print('\n3. 형태소 분석 결과:', text_analyzed)
```

## 옵션 3 - combineN = True
 - 명사+명사로 나오는 단어는 그냥 붙여서 하나의 단어로 반환하라는 옵션
 - 일부러 붙여쓴 거라면 이걸 그대로 받아야 함.
 - 예를 들어, **전문 공문서**에서는 띄어쓰기를 명확히 하면서, 전문용어를 붙여쓰는 경향이 있으므로 이 옵션을 쓰면 좋음.

```python
text_analyzed = rhinoMorph.onlyMorph_list(rn, text, pos=['NNG', 'NNP',
        'VV', 'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], combineN=True)
print('형태소 분서 결과:\n', text_analyzed)
```

```python
text_analyzed = rhinoMorph.wholeResult_list(rn, text, combineN=True)
print('형태소 분석 결과: \n', text_analyzed)
```

## 옵션 4 - xrVv=True

 - 언근 + 하 형태로 하나의 동사로 출력하기
 - 품사태그 XSV(eXtended Suffix of Verve)는 동사파생접미사인데 이를 앞의 어근에 붙여서 출력한다.
 - xrVv 아규먼트로, 기본값은 False로서 둘을 분리하여 출력한다.
 - 분리된 어근이 명사인 경우, 명사로 출력된다.

```python
text_analyzed = rhinoMorph.onlyMorph_list(rn, '사랑합니다')
print('옵션 전: ', text_analyzed)

text_analyzed = rhinoMorph.onlyMorph_list(rn, '사랑합니다', xrVv=True)
print('옵션 후: ', text_analyzed)
```

```python
text_analyzed = rhinoMorph.wholeResult_list(rn, '사랑합니다', xrVv=True)
print(text_analyzed)
```

# 형태소 분석 함수2 - wholeResult_list()
- return 값으로 형태소 분석과 품사태그를 받는다.
- 결과를 하나의 긴 string으로 볼려면, rhinoMorph.wholeResult_text(rn, text) 함수 이용

```python
morphs, poses = rhinoMorph.wholeResult_list(rn, text)
print('형태소 분석 결과:')
print('morphs:', morphs)
print('poses:', poses)
```

```python
text_analyzed = rhinoMorph.wholeResult_text(rn, text)
print('형태소 분석 결과:\n', text_analyzed)
```
